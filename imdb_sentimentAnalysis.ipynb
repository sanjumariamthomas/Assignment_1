```python
import pandas as pd
import numpy as np
import nltk
import string
string.punctuation
from nltk.classify import NaiveBayesClassifier
```

```python
tweets = pd.read_csv('D:/assinmntnlp/Tweets.csv', sep=',',encoding='latin-1')
```

```python
tweets.head(10)
```

```python
words_remove = nltk.corpus.stopwords.words("english") + list(string.punctuation)
```

```python
words_remove[:5]
```

```python
def word_filter(words):
    return {
        word:1 for word in words \
if not word in words_remove}
```

```python
is_pos = tweets['airline_sentiment'].str.contains("positive")
is_neg = tweets['airline_sentiment'].str.contains("negative")
is_neutral = tweets['airline_sentiment'].str.contains("neutral")
```

```python
positive_tweets = tweets[is_pos]
print(len(positive_tweets))
```

```python
negative_tweets = tweets[is_neg]
print(len(negative_tweets))
```

```python
neutral_tweets = tweets[is_neutral]
print(len(neutral_tweets))
```

```python
tokenized_negative_tweets = []
for text in negative_tweets['text']:
        tokenized_negative_tweets.append(nltk.word_tokenize(text))
```

```python
tokenized_negative_tweets[:5]
```

```python
negative_features = [(word_filter(text), 'neg') \
    for text in tokenized_negative_tweets]
print(negative_features)
```

```python
tokenized_positive_tweets = []
for text in positive_tweets['text']:
        tokenized_positive_tweets.append(nltk.word_tokenize(text))
tokenized_positive_tweets
positive_features = [
    (word_filter(text), 'pos') \
    for text in tokenized_positive_tweets]
```

```python
tokenized_neutral_tweets = []
for text in neutral_tweets['text']:
        tokenized_neutral_tweets.append(nltk.word_tokenize(text))
tokenized_neutral_tweets
neutral_features = [
    (word_filter(text), 'neu') \
    for text in tokenized_neutral_tweets]
```

```python
train_set = negative_features[:5500] + positive_features[:2000] + neutral_features[:2500]
test_set =  negative_features[5500:] + positive_features[2000:] + neutral_features[2500:]
print(len(train_set),len(test_set))
```

```python
classifier = NaiveBayesClassifier.train(train_set)
```

```python
accuracy = nltk.classify.util.accuracy(classifier, test_set)
print(accuracy)
```
